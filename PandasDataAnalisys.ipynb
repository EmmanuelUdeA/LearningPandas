Excluding and identify outliers

````python
data = pd.read_csv('residential_rent.csv', sep = ';')
data.head(10)
````

````python
data.boxplot(['Valor'])
````

````python
data[data['Valor'] >= 500000]
````

````python
value = data['Valor']
````

````python
Q1 = value.quantile(.25)
Q3 = value.quantile(.75)
IIQ = Q3 - Q1
lower_limit = Q1 - 1.5 * IIQ
upper_limit = Q3 + 1.5 * IIQ

````

````python
selection = ((value >= lower_limit) & (value <= upper_limit))
````

````python
new_data = data[selection]
````

````python
new_data.boxplot(['Valor'])
````

````python
data.hist(['Valor'])
````

````python
new_data.hist(['Valor'])
````

````python
data.boxplot(['Valor'], by = ['Tipo'])
````

````python
group_type = data.groupby('Tipo')['Valor']
````

````python
Q1 = group_type.quantile(.25)
Q3 = group_type.quantile(.75)
IIQ = Q3 - Q1
lower_limit = Q1 - 1.5 * IIQ
upper_limit = Q3 + 1.5 * IIQ
````

````python
new_data = pd.DataFrame()
for tipo in group_type.groups.keys():
  is_type = data['Tipo'] == tipo
  within_limit = (data['Valor'] >= lower_limit[tipo]) & (data['Valor']<=upper_limit[tipo])
  selection = is_type & within_limit
  selection_data = data[selection]
  new_data = pd.concat([new_data, selection_data])
````

````python
new_data.boxplot(['Valor'], by = ['Tipo'])
````

````python
new_data.to_csv('residential_rent_without_outliers.csv', sep = ';', index = False)
````
